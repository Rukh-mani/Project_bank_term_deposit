```python
#Task 10 is there any effect of class passengers traveled in and survival rate?

titanic_df[['Pclass', 'Survived']]

condition_two = (titanic_df.Survived ==0)

# Filter for rows where passengers did not survive
non_survivors = titanic_df.loc[condition_two, ['Pclass', 'Survived']]

# Display the filtered table
non_survivors

third_class_unalive = (non_survivors['Pclass'] ==3).sum()
second_class_unalive = (non_survivors['Pclass']==2).sum()
first_class_unalive = (non_survivors['Pclass']==1).sum()

print(f'total number of non-survivors is {len(non_survivors)} out of which {(third_class_unalive/len(non_survivors))*100} % is from third class')
```

```python
#chatgpt improved way of printing info:
total_non_survivors = len(non_survivors)
# Display percentage of non-survivors in each class
print(f"Total number of non-survivors: {total_non_survivors}")
print(f"Percentage of non-survivors from third class: {(third_class_unalive / total_non_survivors) * 100:.2f}%")
print(f"Percentage of non-survivors from second class: {(second_class_unalive / total_non_survivors) * 100:.2f}%")
print(f"Percentage of non-survivors from first class: {(first_class_unalive / total_non_survivors) * 100:.2f}%")
```

```python
#chatgpt for visualization

import matplotlib.pyplot as plt

# Prepare data for visualization
non_survivor_counts = [third_class_unalive, second_class_unalive, first_class_unalive]
classes = ['Third Class', 'Second Class', 'First Class']

# Plotting
plt.bar(classes, non_survivor_counts)
plt.title("Non-Survivor Counts by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Non-Survivors")
plt.show()
```

```python
#Task 11 Do missing cabin values have something incommon with Pclass?

missing_cabin_values = titanic_df["Cabin"].isna()  #finding out number of values in "Cabin" that are NaN
missing_cabin_values.sum() #total number of NaN values

#grouping based on class and then getting number of NaN cabins by taking sum of NaNs under each group of 'Pclass'
missing_cabin_by_class = titanic_df.groupby('Pclass')['Cabin'].apply(lambda x: x.isna().sum()) 

print("Number of Missing Cabin Values per Class:")
print(missing_cabin_by_class)
```

```python
#Task 13 group Fare by 3 thresholds so you have category cheap fare, regular fare, expensive fare
minimum_fare = titanic_df.Fare.min()
mid_fare = titanic_df.Fare.median()
max_fare = titanic_df.Fare.max()


# Define bins with an additional threshold for clear separation
titanic_df['fare_category'] = pd.cut(
    titanic_df['Fare'],
    bins=[minimum_fare, mid_fare, (mid_fare + max_fare) / 10, max_fare],
    labels=['cheap fare', 'regular fare', 'expensive fare'],
    include_lowest=True
)

# Display the updated DataFrame
titanic_df

```
