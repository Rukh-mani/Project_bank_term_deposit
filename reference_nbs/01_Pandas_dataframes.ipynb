# Joining and merging

```python
df2 = {"Content Rating":['Everyone', 'Teen', 'Everyone 10+', 'Mature 17+', 'Adults only 18+', 'Unrated', np.nan],
 "Restriction":["No", "No", "Yes", "Yes","Yes", "N/A", "N/A"]}
df2 = pd.DataFrame(df2)

# Merge two DataFrames by a common column
print(len(df))
merged_df = pd.merge(df, df2, on='Content Rating') #creates a new df with an extra column - no.of rows wont change
print(len(merged_df))
print(merged_df[["Content Rating", "Restriction"]].isna().sum()) #find rows that have Nans in content rating and restriction columns
```

```python
merged_df[merged_df["Content Rating"].isna()] 
```

```python
df2 = {"Content Rating":['Everyone', 'Teen', 'Everyone 10+', 'Mature 17+', 'Adults only 18+', 'Unrated'],
 "Restriction":["No", "No", "Yes", "Yes","Yes", "N/A"]}
df2 = pd.DataFrame(df2)

# Merge with different types of joins (e.g., inner, outer, left, right)
merged_df = pd.merge(df, df2, on='Content Rating', how='inner')

print(len(df))
print(len(merged_df))
```

```python
pd.concat([df, df2], axis=0) #concat by rows - top to bottom
```

```python
pd.concat([df, df2], axis=1) #concat by columns - side to side
```

# Transforming columns

```python
def multiply_twice(x):
    return x * 2

df['new_rating_var1'] = df['Rating'].apply(multiply_twice) #passing the function as a reference
df.head()
```

```python
# Apply a lambda function to a column
# lambda is a way for defining the anonymous function
df['new_rating_var2'] = df['Rating'].apply(lambda x: x * 2)


# Using lambda with conditionals
df['rating_conditional'] = df['Rating'].apply(lambda x: x * 2 if x > 1 else x)
```

# Aggregating dataframe

```python
# Calculate mean, sum, and other aggregate functions
mean_value = df['Rating'].mean()
sum_value = df['Rating'].sum()

print(mean_value, sum_value)

# Group by a column and apply aggregation functions
grouped_df = df.groupby('Content Rating')['Rating'].agg(['mean', 'sum', 'count'])
grouped_df
```

```python
# TODO use as_index =  False parameter in group by function to not use Content Rating as index
# grouped_df_new = df.groupby('Content Rating', as_index=False)['Rating']
# print(grouped_df_new)

# perform the grouping by multiple columns e.g. Content Rating and Type

grouped_df_new = df.groupby(['Content Rating', 'Type'], as_index = False)['Rating'].agg(['mean', 'sum', 'count'])
grouped_df_new

#if as_index is false, then the indices are just numbers and content rating or type is not used as the index
```
